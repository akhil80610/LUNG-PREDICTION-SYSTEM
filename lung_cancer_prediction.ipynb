```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#READING DATA

```python
dataset=pd.read_csv('lung cancer survey.csv')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

#REDUCING DUPLICATE ENTRIES

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

#COMPARING PEER PRESSURE AND ANXIETY

```python
sns.countplot(data=dataset,x='PEER_PRESSURE',hue="LUNG_CANCER")
plt.show()
```

```python
sns.countplot(data=dataset,x='ANXIETY',hue="LUNG_CANCER")
plt.show()
```

```python
dataset.drop("PEER_PRESSURE",inplace=True,axis=1)
```

```python
dataset
```

#LABEL ENCODING DATA

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["GENDER"]=le.fit_transform(dataset["GENDER"])
dataset["LUNG_CANCER"]=le.fit_transform(dataset["LUNG_CANCER"])
```

GENDER     -male=1      -female=0

LUNG_CANCER  -yes=1   -no=0

```python
dataset
```

#CHECKING CORRELATION

```python
dataset.drop("GENDER",inplace=True,axis=1)
dataset.drop("SHORTNESS OF BREATH",inplace=True,axis=1)
```

```python
dataset
```

#CLEANING OUTLIERS

```python
dataset.boxplot(column=["AGE"])
```

```python
Q1=dataset["AGE"].quantile(0.25)
Q3=dataset["AGE"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
dataset_cleaned= dataset[(dataset["AGE"] >= lower)& (dataset["AGE"] <= upper)].copy() # Added .copy() to avoid SettingWithCopyWarning

# Remove trailing spaces from column names
dataset_cleaned.columns = dataset_cleaned.columns.str.strip()
```

```python
dataset_cleaned
```

#SCALING DATASET

```python
print(dataset_cleaned.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define features excluding the target variable 'LUNG_CANCER'
features=['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'CHRONIC DISEASE',
          'FATIGUE', 'ALLERGY', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING',
          'SWALLOWING DIFFICULTY', 'CHEST PAIN']

scaler=MinMaxScaler()
# Scale only the feature columns from the cleaned dataset
scaled_data = scaler.fit_transform(dataset_cleaned[features].copy())
scaled_dataset = pd.DataFrame(scaled_data , columns=features)
```

```python
scaled_dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

#SPLITING DATA

```python
# x should be all features (scaled_dataset)
x= scaled_dataset

# y should be the target variable (LUNG_CANCER) from the cleaned dataset
y = dataset_cleaned['LUNG_CANCER']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.columns)
```

#MODEL

MODEL-1 LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy Score =",accuracy)
```

MODEL-2 RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2=RandomForestClassifier()
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy Score =",accuracy)
```

```python
print("Answer the following Questions:\n")
user_data = {}
user_data['AGE'] = int(input("Enter your age: "))
user_data['SMOKING'] = input("do you smoke? (yes/no): ")
user_data['YELLOW_FINGERS'] = input("do you have yellow fingers? (yes/no): ")
user_data['ANXIETY'] = input("do you have anxiety? (yes/no): ")
user_data['CHRONIC DISEASE'] = input("do you any chronic disease? (yes/no): ")
user_data['FATIGUE'] = input("do you have fatigue? (yes/no): ")
user_data['ALLERGY'] = input("do you have any allergies? (yes/no): ")
user_data['WHEEZING'] = input("do you have wheezing? (yes/no): ")
user_data['ALCOHOL CONSUMING'] = input("do you consume alcohol? (yes/no): ")
user_data['COUGHING'] = input("do you having coughing? (yes/no): ")
user_data['SWALLOWING DIFFICULTY'] = input("do you have difficulty swallowing? (yes/no): ")
user_data['CHEST PAIN'] = input("do you have chest pain? (yes/no): ")


features=['AGE' ,'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'CHRONIC DISEASE', 'FATIGUE',
 'ALLERGY', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING',
 'SWALLOWING DIFFICULTY', 'CHEST PAIN']


for key in user_data:
    if isinstance(user_data[key], str):
        if user_data[key].lower() == "yes":
            user_data[key] = 2
        elif user_data[key].lower() == "no":
            user_data[key] = 1
        else:
            print(f"Invalid input for {key}: {user_data[key]}")



data_list = [user_data[feature] for feature in features]


scaled_data = scaler.transform([data_list])
scaled_data_df = pd.DataFrame(scaled_data, columns=features)


prediction_lr = model1.predict(scaled_data_df)
prediction_rf = model2.predict(scaled_data_df)



print("\n--- Prediction Results ---")
print("Based on Logistic Regression Model:")
if prediction_lr[0] == 1:
    print("\nPrediction: High risk of Lung Cancer (1)")
else:
    print("\nPrediction: Low risk of Lung Cancer (0)")

print("\nBased on Random Forest Classifier Model:")
if prediction_rf[0] == 1:
     print("\nPrediction: High risk of Lung Cancer (1)")
else:
     print("\nPrediction: Low risk of Lung Cancer (0)")
```
